```python
# Gerekli kütüphaneleri yükleyin
!pip install git+https://github.com/rongardF/tvdatafeed
!pip install tradingview-screener==2.5.0

import numpy as np
import pandas as pd
from tvDatafeed import TvDatafeed, Interval
import warnings
from tradingview_screener import get_all_symbols

warnings.simplefilter(action='ignore')

# TradingView'e giriş yapmadan bağlanın
tv = TvDatafeed()

# Türkiye piyasasındaki tüm hisse senetlerini otomatik olarak alın
symbols = get_all_symbols(market='turkey')
symbols = [symbol.replace('BIST:', '') for symbol in symbols]
symbols = sorted(symbols)

# Geleneksel pivot seviyelerini hesaplamak için fonksiyon
def traditional_pivots(high, low, close):
    pivot = (high + low + close) / 3
    r1 = 2 * pivot - low
    s1 = 2 * pivot - high
    r2 = pivot + (high - low)
    s2 = pivot - (high - low)
    r3 = pivot + 2 * (high - low)
    s3 = pivot - 2 * (high - low)
    return pivot, r1, s1, r2, s2, r3, s3

# Verileri saklamak için boş bir DataFrame oluşturun
results = []

# Her bir hisse senedi için verileri alın ve geleneksel pivot seviyelerini hesaplayın
for symbol in symbols:
    try:
        # Haftalık veri çekin (son 12 hafta)
        df_weekly = tv.get_hist(symbol=symbol, exchange='BIST', interval=Interval.in_weekly, n_bars=12)

        # Son kapanış verilerini alın (son kapanan bar)
        high = df_weekly['high'].iloc[-1]
        low = df_weekly['low'].iloc[-1]
        close = df_weekly['close'].iloc[-1]
        open_weekly = df_weekly['open'].iloc[-1]

        # Günlük veri çekin (en son kapanış)
        df_daily = tv.get_hist(symbol=symbol, exchange='BIST', interval=Interval.in_daily, n_bars=1)
        current_price = df_daily['close'].iloc[-1]

        # Geleneksel pivot seviyelerini hesaplayın
        pivot, r1, s1, r2, s2, r3, s3 = traditional_pivots(high, low, close)

        # Koşulları kontrol edin
        is_valid = (current_price > pivot) and (current_price > open_weekly) and (current_price < r2)

        # Koşul sağlanıyorsa anlık olarak ekrana yazdır ve tabloya ekle
        if is_valid:
            result = [symbol, current_price, pivot, r2, open_weekly, is_valid]
            results.append(result)

            # Anlık sonuçları ekrana yazdır
            print(f"Koşulları Sağlayan Hisse: {symbol}, Fiyat: {current_price}, Pivot: {pivot}, R2: {r2}, Haftalık Açılış: {open_weekly}")

    except Exception as e:
        print(f"Error processing {symbol}: {e}")

# Sonuçları DataFrame'e dönüştürün
df_results = pd.DataFrame(results, columns=['Symbol', 'Current Price', 'Pivot', 'R2', 'Weekly Open', 'Condition Met'])

# Sonuçları Excel dosyasına kaydedin
excel_filename = 'filtered_stocks.xlsx'
df_results.to_excel(excel_filename, index=False)

print(f"Sonuçlar '{excel_filename}' dosyasına kaydedildi.")
```
